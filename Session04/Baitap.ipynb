```python
# Bài 5
scores = {"An": 8, "Bình": 9, "Châu": 7}

# In danh sách khóa 
print("Danh sách tên học sinh:", list(scores.keys()))

# In danh sách giá trị
print("Danh sách điểm số:", list(scores.values()))

# In từng cặp khóa-giá trị
for name, score in scores.items():
    print(f"Tên: {name}. Điểm: {score}")
```

```python
# Bài 7
products = {"A": 120, "B": 80, "C": 300, "D": 50}

# Xóa tất cả sản phẩm có giá dưới 100
products = {k: v for k, v in products.items() if v >= 100} 

# In danh sách còn lại
print("Sản phẩm còn lại sau khi xóa:", products)

# Thêm 1 sản phẩm và in ra toàn bộ danh sách 
products["E"] = 150
print("Danh sách sản phẩm sau khi thêm:", products)

# Đếm số lượng các sản phẩm
print(f"Số lượng sản phẩm: {len(products)}")
```

```python
# Bài 8
products = {"A": 120, "B": 80, "C": 300, "D": 50}

# Kiểm tra sản phẩm A
if "A" in products:
    print(f"Giá của sản phẩm A là: {products['A']}")
else:
    print("Sản phẩm A đã bị xóa")

# Sắp xếp giá từ thấp đến cao
sorted_products = dict(sorted(products.items(), key=lambda item: item[1]))
print("Sản phẩm sắp xếp theo giá từ thấp đến cao:", sorted_products)

# Tổng giá sản phẩm
print(f"Tổng giá của tất cả sản phẩm là: {sum(products.values())}")

# Xóa sản phầm có giá > 200
products = {k: v for k, v in products.items() if v <= 200}
print("Sản phẩm sau khi xóa sản phẩm có giá > 200:", products)
```
