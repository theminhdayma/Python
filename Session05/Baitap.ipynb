```python
#Bài 2

import csv

with open('students.csv', mode='r', encoding='utf-8') as infile:
    reader = csv.reader(infile)
    header = next(reader)
    
    high_grade_students = []

    for row in reader:
        name, age, grade = row[0], row[1], float(row[2])
        if grade >= 7:
            high_grade_students.append(row)
            print(f"Sinh viên: {name}, Tuổi: {age}, Điểm: {grade}")

with open('high_grade_students.csv', mode='w', newline='', encoding='utf-8') as outfile:
    writer = csv.writer(outfile)
    writer.writerow(header) 
    writer.writerows(high_grade_students) 
```

```python
# Bài 3

import json

with open('user_data.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

person = next((item for item in data if item.get('id') == 1), None)
if person:
    print(f"Tên: {person.get('name')}, Tuổi: {person.get('age')}")
else:
    print("Không tìm thấy người có id = 1")
new_user = {
    "id": 2,
    "name": "Nguyen Van B",
    "age": 28
}

data.append(new_user)

with open('user_data.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```

```python
# Bài 6

import json

file_path = 'inventory.json'

with open(file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

new_product = {
    "product_id": 101,
    "name": "Sản phẩm mới",
    "quantity": 50,
    "price": 120000
}

data.append(new_product)

with open(file_path, 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print("Đã thêm sản phẩm mới và cập nhật file JSON thành công.")
```

```python
# Bài 8

from datetime import datetime
from pathlib import Path

today = datetime.today().date()
log_file = Path("log.txt")

with log_file.open("r", encoding="utf-8") as file:
    for line in file:
        parts = line.strip().split(" - ", 1)
        if len(parts) != 2:
            continue 
        
        timestamp_str, message = parts

        try:
            timestamp = datetime.strptime(timestamp_str, "%Y-%m-%d %H:%M:%S")
        except ValueError:
            continue

        if timestamp.date() == today:
            print(line.strip())
```
