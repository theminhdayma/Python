```python
# Bài 3 

import pandas as pd
import numpy as np

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, None, 22, 20]
score = [8.5, None, 6.5, 9.0]

df = pd.DataFrame({'name': name, 'age': age, 'score': score})

print(df.isna())

mean_age = df['age'].mean()
df['age'] = df['age'].fillna(mean_age)
print(df)

mean_score = df['score'].mean()
df['score'] = df['score'].fillna(mean_score)
print(df)

df_dropna = df.dropna()
print(df_dropna)

```

```python
# Bài 4

import pandas as pd

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, 21, 22, 20]
score = [8.5, 7.0, 6.5, 9.0]
df = pd.DataFrame({'name': name, 'age': age, 'score': score})

df['age_group'] = df['age'].apply(lambda x: 'Young' if x <= 20 else 'Adult')
print(df)

score_map = {8.5: 'Excellent', 7.0: 'Good', 6.5: 'Average', 9.0: 'Excellent'}
df['score_level'] = df['score'].map(score_map)
print(df)

df['score'] = df['score'].astype(int)
print(df[['name', 'age', 'score']])

df['name'] = df['name'].replace('Bình', 'Bình (Updated)')
print(df[['name', 'age', 'score']])

```

```python
# Bài 5

import pandas as pd

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, 21, 22, 20]
score = [8.5, 7.0, 6.5, 9.0]
df = pd.DataFrame({'name': name, 'age': age, 'score': score})

grouped = df.groupby('age')['score'].agg(
    mean_score='mean',
    min_score='min',
    max_score='max'
).reset_index()
print(grouped)

```

```python
import os
import glob
import json
import pandas as pd

# Đường dẫn tới thư mục log
log_dir = "logs/"
log_files = glob.glob(os.path.join(log_dir, "*.json"))  # Đọc mọi file JSON

report = []

for log_file in log_files:
    # Trích xuất ngày từ tên file (giả sử đúng cấu trúc log_YYYY-MM-DD.json)
    basename = os.path.basename(log_file)
    try:
        date_part = basename.split('_')[1].split('.')[0]
        date = pd.to_datetime(date_part)
    except Exception as ex:
        print(f"Lỗi lấy ngày từ file {basename}: {ex}")
        continue

    # Đọc file JSON (dòng đơn hoặc danh sách dict)
    with open(log_file, 'r', encoding='utf-8') as f:
        try:
            data = json.load(f)
        except Exception as ex:
            print(f"Lỗi đọc file JSON {basename}: {ex}")
            continue

        # Nếu file là dict, lấy trực tiếp
        if isinstance(data, dict):
            duration = data.get('duration')
            if duration is not None:
                # Có thể thêm cả user/action nếu muốn
                report.append({
                    'date': date,
                    'user': data.get('user'),
                    'action': data.get('action'),
                    'duration': duration
                })
        # Nếu file là list, duyệt từng phần tử
        elif isinstance(data, list):
            for entry in data:
                duration = entry.get('duration')
                if duration is not None:
                    report.append({
                        'date': date,
                        'user': entry.get('user'),
                        'action': entry.get('action'),
                        'duration': duration
                    })
        else:
            print(f"Định dạng JSON không đúng: {basename}")

# Xây DataFrame từ dữ liệu đã collect
df = pd.DataFrame(report)
print("Kiểm tra DataFrame tổng hợp:")
print(df)

# Tổng hợp duration theo ngày (nếu muốn tổng hợp theo user, thêm groupby theo user)
summary = df.groupby('date')['duration'].sum().reset_index()

# Xuất ra file CSV báo cáo
summary.to_csv('log_daily_duration.csv', index=False)

print("\nTổng duration theo ngày:")
print(summary)
```
