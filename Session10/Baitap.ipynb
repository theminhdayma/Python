```python
# Bài 1

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)  
scores = np.random.normal(7, 1.5, 200)

plt.figure(figsize=(8, 4))
sns.histplot(scores, bins=15, kde=True, color='skyblue')

plt.title("Phân phối điểm thi")
plt.xlabel("Điểm")
plt.ylabel("Tần suất")

plt.show()
```

```python
# Bài 2

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
subjects = ['Toán', 'Lý', 'Hóa']
n_per_subj = 100

data = {
    "subject": np.repeat(subjects, n_per_subj),
    "score": np.concatenate([
        np.random.normal(7.5, 1.0, n_per_subj),
        np.random.normal(6.5, 1.2, n_per_subj),
        np.random.normal(6.0, 1.3, n_per_subj),
    ])
}

df = pd.DataFrame(data)

sns.displot(
    df,
    x="score",
    hue="subject",
    kind="kde",
    fill=False,
    height=4,
    aspect=1.5
)

plt.title("Phân phối điểm theo môn học")
plt.xlabel("Điểm")
plt.ylabel("Mật độ")
plt.xlim(0, 10)
plt.show()
```

```python
# Bài 3

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
hours = np.arange(1, 11)  
noise = np.random.normal(0, 0.8, size=hours.size)
score = hours * 1.2 + noise    

plt.figure(figsize=(6, 4))
plt.scatter(hours, score, color='blue')
plt.title("Mối quan hệ giữa giờ học và điểm thi")
plt.xlabel("Số giờ học")
plt.ylabel("Điểm thi")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
# Bài 4

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
classes = np.random.choice(['A', 'B', 'C'], size=60, p=[0.3, 0.3, 0.4])
scores = np.random.normal(7, 1, size=60)  

df = pd.DataFrame({
    "class": classes,
    "score": scores
})

plt.figure(figsize=(6, 4))
sns.barplot(data=df, x="class", y="score", estimator="mean", ci=95)
plt.title("Điểm trung bình theo lớp")
plt.xlabel("Lớp")
plt.ylabel("Điểm trung bình")
plt.ylim(0, 10)
plt.show()
```

```python
# Bài 5

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)

majors = np.repeat(['CNTT', 'Kinh Tế', 'Ngôn Ngữ'], 80)
scores = np.concatenate([
    np.random.normal(7.8, 0.8, 80), 
    np.random.normal(7.0, 1.0, 80),  
    np.random.normal(6.5, 1.2, 80),  
])

df = pd.DataFrame({"major": majors, "score": scores})

plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x="major", y="score")
plt.title("Boxplot điểm theo nhóm ngành")
plt.xlabel("Ngành")
plt.ylabel("Điểm")
plt.ylim(0, 10)
plt.show()

plt.figure(figsize=(8, 4))
sns.violinplot(data=df, x="major", y="score")
plt.title("Violinplot điểm theo nhóm ngành")
plt.xlabel("Ngành")
plt.ylabel("Điểm")
plt.ylim(0, 10)
plt.show()
```

```python
# Bài 6

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Tạo dữ liệu ngày và doanh thu
np.random.seed(0)

day = np.arange(1, 31)
base_trend = day * 5                     # xu hướng tăng dần
noise = np.random.normal(0, 20, size=day.size)  # nhiễu
sales = base_trend + noise

df = pd.DataFrame({"day": day, "sales": sales})

# 2. Lineplot: doanh thu theo ngày
plt.figure(figsize=(8, 4))
sns.lineplot(data=df, x="day", y="sales", marker="o")
plt.title("Doanh thu theo ngày trong tháng")
plt.xlabel("Ngày")
plt.ylabel("Doanh thu")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

# 3. Regplot: scatter + đường hồi quy tuyến tính
plt.figure(figsize=(8, 4))
sns.regplot(data=df, x="day", y="sales")
plt.title("Xu hướng doanh thu (regression)")
plt.xlabel("Ngày")
plt.ylabel("Doanh thu")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
# Bài 7

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)

n = 500
hours = np.random.uniform(1, 10, n)              
noise = np.random.normal(0, 1.5, n)     
score = hours * 1.5 + 4 + noise                   

majors = np.random.choice(['A', 'B', 'C'], size=n) 

df = pd.DataFrame({
    "hours": hours,
    "score": score,
    "major": majors
})

plt.figure(figsize=(7, 4))
sns.histplot(df["score"], bins=20, kde=True)
plt.title("Phân phối điểm thi")
plt.xlabel("Điểm")
plt.ylabel("Tần suất")
plt.show()

plt.figure(figsize=(7, 4))
sns.scatterplot(data=df, x="hours", y="score", alpha=0.7)
plt.title("Mối quan hệ giữa giờ học và điểm")
plt.xlabel("Giờ học")
plt.ylabel("Điểm")
plt.grid(True, linestyle="--", alpha=0.4)
plt.show()

plt.figure(figsize=(7, 4))
sns.boxplot(data=df, x="major", y="score")
plt.title("Phân phối điểm theo nhóm ngành")
plt.xlabel("Nhóm ngành")
plt.ylabel("Điểm")
plt.show()
```

```python
# Bài 8

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)

months = np.tile(np.arange(1, 13), 3) 
categories = np.repeat(['Shoes', 'Clothes', 'Bags'], 12)

base = {
    'Shoes':   80,
    'Clothes': 60,
    'Bags':    50
}

trend = {
    'Shoes':   5, 
    'Clothes': 3,
    'Bags':    2
}

revenues = []
for m, c in zip(months, categories):
    noise = np.random.normal(0, 10)       
    rev = base[c] + trend[c] * m + noise
    revenues.append(rev)

df = pd.DataFrame({
    'month': months,
    'category': categories,
    'revenue': revenues
})

plt.figure(figsize=(8, 4))
sns.lineplot(data=df, x='month', y='revenue', hue='category', marker='o')
plt.title('Doanh thu theo tháng theo nhóm sản phẩm')
plt.xlabel('Tháng')
plt.ylabel('Doanh thu')
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()

plt.figure(figsize=(8, 4))
sns.violinplot(data=df, x='category', y='revenue')
plt.title('Phân phối doanh thu theo nhóm sản phẩm')
plt.xlabel('Nhóm sản phẩm')
plt.ylabel('Doanh thu')
plt.show()

plt.figure(figsize=(8, 4))
sns.regplot(data=df, x='month', y='revenue')
plt.title('Xu hướng doanh thu toàn bộ sản phẩm')
plt.xlabel('Tháng')
plt.ylabel('Doanh thu')
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```
