```python
# Bài 1

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 20)
y = x**2 + 2*x

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y, 'b-', linewidth=2, marker='o', markersize=4, label='y = x² + 2x')

ax.set_title('Biểu đồ của Nguyễn Thế Minh')

ax.set_xlabel('Trục x')
ax.set_ylabel('Trục y')

ax.grid(True, linestyle='--', alpha=0.7)

ax.legend()

plt.show()
```

```python
# Bài 4

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 100)

y1 = x        
y2 = x**2    
y3 = x**3  

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y1, 'b-', linewidth=2, label='y = x', marker='o', markersize=3)
ax.plot(x, y2, 'r-', linewidth=2, label='y = x²', marker='s', markersize=3)
ax.plot(x, y3, 'g-', linewidth=2, label='y = x³', marker='^', markersize=3)

ax.set_title('So sánh ba hàm: y = x, y = x², y = x³')
ax.set_xlabel('Trục x')
ax.set_ylabel('Trục y')

ax.legend()

ax.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python
# Bài 5

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 50)
y1 = x        
y2 = x**2    

x_rand = np.random.uniform(0, 10, 50)
y_rand = np.random.uniform(0, 50, 50)

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(x, y1, 'b-', label='y = x')
axs[0, 0].set_title('Đường thẳng y = x')
axs[0, 0].set_xlabel('Trục x')
axs[0, 0].set_ylabel('Trục y')
axs[0, 0].grid(True, linestyle='--', alpha=0.7)
axs[0, 0].legend()

axs[1, 0].plot(x, y2, 'r-', label='y = x²')
axs[1, 0].set_title('Đường cong y = x²')
axs[1, 0].set_xlabel('Trục x')
axs[1, 0].set_ylabel('Trục y')
axs[1, 0].grid(True, linestyle='--', alpha=0.7)
axs[1, 0].legend()

axs[0, 1].scatter(x_rand, y_rand, color='green', label='Điểm ngẫu nhiên')
axs[0, 1].set_title('Scatterplot ngẫu nhiên')
axs[0, 1].set_xlabel('Trục x')
axs[0, 1].set_ylabel('Trục y')
axs[0, 1].grid(True, linestyle='--', alpha=0.7)
axs[0, 1].legend()

axs[1, 1].scatter(y_rand, x_rand, color='purple', label='Điểm ngẫu nhiên đảo')
axs[1, 1].set_title('Scatterplot ngẫu nhiên đảo trục')
axs[1, 1].set_xlabel('Trục x')
axs[1, 1].set_ylabel('Trục y')
axs[1, 1].grid(True, linestyle='--', alpha=0.7)
axs[1, 1].legend()

plt.tight_layout()

plt.show()
```

```python
# Bài 6

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y, label='y = sin(x)', color='blue')

ax.annotate('Điểm đặc biệt\n(x = π/2, y = 1)', 
            xy=(np.pi/2, 1), xycoords='data',
            xytext=(np.pi/2 + 1, 0.5), textcoords='data',
            arrowprops=dict(facecolor='black', arrowstyle='->'),
            fontsize=12, fontweight='bold')

ax.set_title('Đường y = sin(x)')
ax.set_xlabel('Trục x')
ax.set_ylabel('Trục y')

ax.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

x = np.linspace(0, 2 * np.pi, 100)
fig, ax = plt.subplots(figsize=(10, 6))
line, = ax.plot(x, np.sin(x), color='blue')

ax.set_ylim(-1.5, 1.5)
ax.set_title('Đồ thị động y = sin(x + t)')
ax.set_xlabel('Trục x')
ax.set_ylabel('Trục y')
ax.grid(True, linestyle='--', alpha=0.7)

def update(frame):
    y = np.sin(x + 0.1 * frame)
    line.set_ydata(y)
    return line,

ani = FuncAnimation(fig, update, frames=200, interval=30, blit=True, repeat=True)

plt.show()
```

```python
# Bài 9

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime, timedelta

start_date = datetime(2025, 12, 1)
dates = [start_date + timedelta(days=i) for i in range(30)]
np.random.seed(42) 

trend = np.linspace(20, 28, 30) 
noise = np.random.normal(0, 2, 30) 
temperature = trend + noise

temperature[5] += 8  
temperature[15] -= 10
temperature[25] += 6 

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(dates, temperature, 'b-', linewidth=2, marker='o', markersize=4, label='Nhiệt độ')

z = np.polyfit(range(30), temperature, 1)
trendline = np.polyval(z, range(30))
ax.plot(dates, trendline, 'r--', linewidth=2, label='Xu hướng')

special_days = [5, 15, 25]
special_labels = ['Nóng bất thường', 'Lạnh đột ngột', 'Nóng đột ngột']
colors = ['orange', 'cyan', 'magenta']

for i, (day, label, color) in enumerate(zip(special_days, special_labels, colors)):
    ax.scatter(dates[day], temperature[day], s=200, c=color, zorder=5)
    ax.annotate(label, xy=(dates[day], temperature[day]), 
                xytext=(10, 10), textcoords='offset points',
                bbox=dict(boxstyle='round,pad=0.3', facecolor=color, alpha=0.3),
                arrowprops=dict(arrowstyle='->', color=color),
                fontsize=10, fontweight='bold')

ax.set_title('Chuỗi thời gian nhiệt độ hàng ngày - Tháng 12/2025', fontsize=14, fontweight='bold')
ax.set_xlabel('Ngày')
ax.set_ylabel('Nhiệt độ (°C)')
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend()
ax.tick_params(axis='x', rotation=45)

plt.setp(ax.xaxis.get_majorticklabels(), ha='right')

plt.tight_layout()
plt.show()
```
